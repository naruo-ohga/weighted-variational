```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import example_Ising_loader as loader


params = {'mathtext.fontset' : 'stix',
          'font.family'      : 'sans-serif',
          'font.sans-serif':['Helvetica'],
          'font.size'        : 9,   
          'text.usetex'      : True,
          'text.latex.preamble':r"""
                \usepackage{type1cm}
                \usepackage{amssymb}
                \usepackage{amsmath}
                \usepackage[varg]{txfonts}
                \usepackage{newtxtext}
                \usepackage{bm}
                \usepackage{helvet}
                \usepackage{sansmath}
                \sansmath
                \newcommand{\nosans}[1]{\unsansmath #1\sansmath}
                """,
        "lines.linewidth": 1.0, # default=1.5
        "axes.labelpad": 1.5, # default=4.0
        "xtick.major.pad": 2.0, # default=3.5
        "ytick.major.pad": 2.0, # default=3.5
        "xtick.bottom": True,
        "ytick.left": True}
plt.rcParams.update(params)


mc = {"red": "tomato",
      "blue": "dodgerblue",
      "green": 'mediumseagreen'}
```

```python
# ===== Parameters =====
parent_folder = 'data-main/'
K_list = loader.K_list
driving_name = ''
N = 12
i = 100 # index of lambda

M = N if driving_name == '' else np.nan

setup_list = [('ferromagnetic', 0), ('Gaussian', 9)]


# ===== Load data =====
df_all = []
for Hamiltonian_name, seed_idx in setup_list:
    folder_name = loader.get_folder_name(Hamiltonian_name, driving_name, seed_idx, parent_folder)
    df = pd.read_csv(f'{folder_name}/example_Ising_partial-action_N={N}_i={i}.txt', sep = "\t")

    df_all.append(df)
    display(df)
```

```python
# Plot
aspect_ratio = 2.8/3.4
fig, ax_all = plt.subplots(2, 1, figsize=(3.4, 3.4*aspect_ratio))

cmap_K_norm = mpl.colors.Normalize(vmin=np.min(K_list), vmax=np.max(K_list)+1)
cmap_K = mpl.cm.ScalarMappable(norm=cmap_K_norm, cmap=mpl.cm.plasma).to_rgba

# ===== Plot deviation =====
for ax, df in zip(ax_all, df_all):
    # partial action
    df_partial = df[df['kind'] == 'partial']
    K = df_partial['K'].values
    energy = df_partial['energy'].values
    deviation = df_partial['deviation'].values
    ax.scatter(energy, deviation, facecolor=cmap_K(K), edgecolor='none', s=.7, rasterized=True)

    # results from the weighted variational method
    df_total = df[df['kind'] == 'total']
    for key, row in df_total.iterrows():
        K = row['K']
        deviation = row['deviation']
        ax.axhline(deviation, color=cmap_K(K), lw=1, ls='--', zorder=-4)


ax_all[1].set_xlabel(r"Energy \nosans{$\epsilon_n$}")
fig.supylabel(r"Deviation from \nosans{$\bm{\alpha}^{(\infty)}$}", fontsize=9, x=0.007, y=0.55)

ax_all[0].set_title(r'\nosans{$\mathrm{(a)}$}', loc='left', x=0.01, y=0.97, va='top', ha='left', fontsize=9, pad=0)
ax_all[1].set_title(r'\nosans{$\mathrm{(b)}$}', loc='left', x=0.01, y=0.97, va='top', ha='left', fontsize=9, pad=0)
ax_all[0].set_ylim(0, 50)
ax_all[1].set_ylim(0, 50)


# ===== Adjust the space =====
fig.subplots_adjust(left=0.11, right=0.995, top=0.995, bottom=0.12, wspace=0.26, hspace=0.20)
# ax1.add_patch(mpl.patches.Rectangle((0.0, 0.0), width=1.0, height=1.0, color="red", fill=False, transform=fig.transFigure, clip_on=False))
# fig.align_labels([ax1, ax2])  

# Add legends
x0, y0, w0, h0 = ax_all[0].get_position().bounds
ax1a = fig.add_axes([x0 + 0.06 + 0.027*0.8, y0+h0 - 0.027*1.8/aspect_ratio, 0.027*5, 0.027/aspect_ratio])
ax1a.imshow([K_list[1:]], cmap=mpl.cm.plasma, norm=cmap_K_norm, aspect='equal') #, extent=[0, len(K_list), 0, 1]
ax1a.set_xticks(np.arange(len(K_list[1:])))
ax1a.set_xticklabels(K_list[1:])
ax1a.set_yticks([])
ax1a.set_xlabel(r'Degree \nosans{$K$}')


if setup_list == [('ferromagnetic', 0), ('Gaussian', 9)]:
    ax_all[0].set_ylim(-58*0.04, 58)
    ax_all[1].set_ylim(-25*0.04, 25)
    fig.savefig(f'{parent_folder}example_Ising_deviation_N={N}_i={i}.pdf', dpi=1200);
```
